```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
#import urllib

a=pd.read_csv('spam.csv',encoding='latin-1',sep='\t')
b=pd.read_table('spam.csv')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vector=TfidfVectorizer()
```

```python
x=a.Message
y=a.Type
```

```python
x_t=vector.fit_transform(x)

#y_t=vector.fit_transform(y)
```

```python
vector.vocabulary_
```

```python
x_tra=x_t.toarray()
#y_tra=y_t
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_tra,y)
```

```python
nbG=GaussianNB()
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
nbG.fit(x_train,y_train)
```

```python
y_pre=nbG.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pre)
```
